```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_heart = pd.read_csv('heart.csv')
df_heart.head()
```

```python
# Change the cholesterol level of above 200 to be 1 (risky amount) and 0 (healthy amount)

df_heart.chol = (df_heart.chol > 200).astype(int)
df_heart.chol.value_counts()
```

```python
x = df_heart.drop('chol', axis=1)
y = df_heart.chol
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
# Logistric Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

lrc = LogisticRegression()
lrc.fit(x_train, y_train)
lrc_pred = lrc.predict(x_test)

print(f'Logistic Regression Accuracy: {accuracy_score(y_test.values, lrc_pred) * 100}%')

conf_matrix = confusion_matrix(y_test.values, lrc_pred)
classification_rep = classification_report(y_test.values, lrc_pred)

print (f'Confusion Matrix: \n{conf_matrix}')
print (f'Classification Report: \n{classification_rep}')

```

```python
# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

dtc = DecisionTreeClassifier(random_state=123).fit(x_train, y_train)
dtc_pred = dtc.predict(x_test)

print(f'Decision Tree Classifier Accuracy: {accuracy_score(y_test.values, dtc_pred) * 100}%')

conf_matrix = confusion_matrix(y_test.values, dtc_pred)
classification_rep = classification_report(y_test.values, dtc_pred)

print (f'Confusion Matrix: \n{conf_matrix}')
print (f'Classification Report: \n{classification_rep}')
```

The Logistic Regression performs better than the Decision Tree Classifier with an accuracy that is about 11.47% higher (83.6% - 72.13%)

This suggests that the logistic regression is able to capture more patterns in the data that lead to correct classification of cholesterol level.
